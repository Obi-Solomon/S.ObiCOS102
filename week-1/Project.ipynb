```python
#Simple Interest
#START -> INPUT (P,R,T) -> COMPUTE I = P(1 + RT/100) -> PRINT I -> END
P = float(input("Enter Principal (P): "))
R = float(input("Enter Interest Rate (R %): "))
T = float(input("Enter Time (T in years): "))

I = P * (1 + (R * T) / 100)

print("Interest (I) =", I)

#Compound Interest
#START -> INPUT (P,R,n,t) -> COMPUTE I = P(1 + R/(n*100))^(n*t) -> PRINT I -> END
P = float(input("Enter Principal (P): "))
R = float(input("Enter Rate (R %): "))
n = int(input("Enter number of compounds per year (n): "))
t = float(input("Enter time (t in years): "))

I = P * (1 + R / (n * 100)) ** (n * t)

print("Interest (I) =", I)

#Annuity Plan
#START -> INPUT (PMT,R,n,t) -> COMPUTE RATE -> COMPUTE A = PMT * [((1+rate)^(n*t) -1) / rate] -> PRINT A -> END
PMT = float(input("Enter periodic payment (PMT): "))
R = float(input("Enter annual interest rate (R %): "))
n = int(input("Enter number of compounding periods per year (n): "))
t = float(input("Enter number of years (t): "))

rate = R / (n * 100)
A = PMT * ((1 + rate) ** (n * t) - 1) / rate

print("Future Value of Annuity (A) =", A)
```
