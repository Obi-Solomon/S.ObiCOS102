```python
#QUADRATIC
#INPUT a, b, c
#COMPUTE discriminant = b^2 - 4ac
#IF discriminant > 0 THEN
#root1 = (-b + sqrt(discriminant)) / (2a)
#root2 = (-b - sqrt(discriminant)) / (2a)
#ELSE IF discriminant == 0 THEN
#root1 = root2 = -b / (2a)
#ELSE
#root1 = (-b + i*sqrt(|discriminant|)) / (2a)
#root2 = (-b - i*sqrt(|discriminant|)) / (2a)
#ENDIF
#OUTPUT root1, root2

#CUBIC
#INPUT A, B, C, D
#USE mathematical cubic formula to compute roots
#OUTPUT root1, root2, root3
#END

#QUARTIC
#INPUT A, B, C, D, E
#USE mathematical quartic formula to compute roots
#OUTPUT root1, root2, root4

#code
import numpy as np
import cmath

print("1-Quadratic  2-Cubic  3-Quartic")
choice = input("Choose: ")

if choice == "1":
    a = float(input("a: "))
    b = float(input("b: "))
    c = float(input("c: "))
    d = b**2 - 4*a*c
    x1 = (-b + cmath.sqrt(d)) / (2*a)
    x2 = (-b - cmath.sqrt(d)) / (2*a)
    print("Roots:", x1, x2)

elif choice == "2":
    A = float(input("A: "))
    B = float(input("B: "))
    C = float(input("C: "))
    D = float(input("D: "))
    print("Roots:", np.roots([A, B, C, D]))

elif choice == "3":
    a = float(input("a: "))
    b = float(input("b: "))
    c = float(input("c: "))
    d = float(input("d: "))
    e = float(input("e: "))
    print("Roots:", np.roots([a, b, c, d, e]))

else:
    print("Invalid choice")
```
