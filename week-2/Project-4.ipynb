```python
# Program to calculate Annual Tax Revenue (ATR) based on years of experience and age

# Get input from user
years = int(input("Enter years of experience: "))
age = int(input("Enter age: "))

# Determine ATR based on conditions
if years > 25 and age >= 55:
    atr = 5600000
elif years > 20 and age >= 45:
    atr = 4480000
elif years > 10 and age >= 35:
    atr = 1500000
else:
    atr = 550000

# Output the result
print(f"The Annual Tax Revenue (ATR) is N{atr:,}")
```
